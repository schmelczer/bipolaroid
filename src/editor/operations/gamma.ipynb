```python
import numpy as np
import matplotlib.pyplot as plt
from gamma import get_random_gamma


all = np.array([])
for i in range(100000):
    img = np.power(np.linspace(0, 1, 256), get_random_gamma()) / 128
    all = np.append(all, np.sum(img))

hist, bin_edges = np.histogram(all, bins=100)
plt.figure(figsize=(5, 3))
plt.plot(bin_edges[1:], list(hist))
plt.grid(True)
```
