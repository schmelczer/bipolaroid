```python
import matplotlib.pyplot as plt
import numpy as np

from editor.utils import get_random_saturation_per_hue_lut


plt.figure(figsize=(5, 3))
plt.plot(
    np.linspace(0, 256, 256),
    get_random_saturation_per_hue_lut(),
    markersize=3,
)
plt.xlabel("t parameter")
plt.ylabel("Interpolated value")
plt.grid(True)
```

```python
from editor.utils import get_random_brightness_lut


plt.figure(figsize=(5, 3))
plt.plot(
    np.linspace(0, 256, 256),
    get_random_brightness_lut(),
    markersize=3,
)
plt.ylim(0, 255)
plt.xlabel("t parameter")
plt.ylabel("Interpolated value")
plt.grid(True)
```
