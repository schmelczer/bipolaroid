```python
import matplotlib.pyplot as plt
import numpy as np
from editor.utils import get_colour_lut

t_values = np.linspace(0, 256, 256)
interpolated_values = get_colour_lut(variance=0.3, type="cubic")

plt.figure(figsize=(10, 6))
plt.plot(
    t_values,
    interpolated_values,
    label="Interpolated Function",
    marker="o",
    markersize=3,
)
plt.xlabel("t parameter")
plt.ylabel("Interpolated value")
plt.grid(True)
plt.legend()
plt.show()
```
